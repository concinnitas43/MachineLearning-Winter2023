# **Iris Classification (kNN)**

## $k$-NN
$k$-Nearest Neighbor => 선택된 점으로부터 거리가 가까운 다른 기존의 데이터들이 그것에 대해 투표를 해서 가장 많이 나온 타겟 값으로 분류한다.

Importing and Initialization

```python
# Importing 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris_dataset = load_iris()

print(iris_dataset.keys())
```

```python
print(iris_dataset['target_names']) # Classify할 목표?
print(iris_dataset['feature_names']) # Attributes
```

```python
print(iris_dataset['target'])
print(type(iris_dataset['target']))
print(iris_dataset['target'].shape)
```

# Sklearn의 train_test_split 

전체 데이터를 training data /  test data 나눔

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# KNN

model.fit함수를 통해 학습 (training data), 이후 score를 통해서 얼마나 잘 수행하였는지 측정 (testing data)

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train, y_train)

print(model.score(X_test, y_test))

print(iris_dataset['target_names'][model.predict(np.array([[5, 2.9, 1, 0.2]]))])
```
